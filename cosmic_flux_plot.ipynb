```python
from scipy.ndimage import gaussian_filter1d

def smooth_with_gaussian(x, y, sigma=1):
    """
    Apply Gaussian kernel smoothing to the data.
    
    Parameters:
    - x: Array of x-values.
    - y: Array of y-values.
    - sigma: The standard deviation of the Gaussian kernel.
    
    Returns:
    - y_smooth: The smoothed y-values.
    """
    y_smooth = gaussian_filter1d(y, sigma=sigma)
    #y_smooth = y
    return x, y_smooth


DM_Linden_x, DM_Linden_y = smooth_with_gaussian(DM_Linden_x, DM_Linden_y)
DM_coogan_x, DM_coogan_y = smooth_with_gaussian(DM_coogan_x, DM_coogan_y)
DM_Korsmeier_upper_x, DM_Korsmeier_upper_y = smooth_with_gaussian(DM_Korsmeier_upper_x, DM_Korsmeier_upper_y)
DM_Korsmeier_lower_x, DM_Korsmeier_lower_y = smooth_with_gaussian(DM_Korsmeier_lower_x, DM_Korsmeier_lower_y)
DM_Nan_Li_upper_x, DM_Nan_Li_upper_y = smooth_with_gaussian(DM_Nan_Li_upper_x, DM_Nan_Li_upper_y)
DM_Nan_Li_lower_x, DM_Nan_Li_lower_y = smooth_with_gaussian(DM_Nan_Li_lower_x, DM_Nan_Li_lower_y)
DM_Nan_Li_x, DM_Nan_Li_y = smooth_with_gaussian(DM_Nan_Li_x, DM_Nan_Li_y)
DM_MKachelrieβ_x, DM_MKachelrieβ_y = smooth_with_gaussian(DM_MKachelrieβ_x, DM_MKachelrieβ_y)
DM_M_Winkler_x, DM_M_Winkler_y = smooth_with_gaussian(DM_M_Winkler_x, DM_M_Winkler_y)
BKG_Blum_x, BKG_Blum_y = smooth_with_gaussian(BKG_Blum_x, BKG_Blum_y)
BKG_Poulin_x, BKG_Poulin_y = smooth_with_gaussian(BKG_Poulin_x, BKG_Poulin_y)
BKG_A_Shukla_x, BKG_A_Shukla_y = smooth_with_gaussian(BKG_A_Shukla_x, BKG_A_Shukla_y)
BKG_M_Kachelrie_background_x, BKG_M_Kachelrie_background_y = smooth_with_gaussian(BKG_M_Kachelrie_background_x, BKG_M_Kachelrie_background_y)


DM_pbar_x, DM_pbar_y = smooth_with_gaussian(DM_pbar_x, DM_pbar_y)
DM_dbar_x, DM_dbar_y = smooth_with_gaussian(DM_dbar_x, DM_dbar_y)
```

```python
plt.xscale('log')
plt.yscale('log')
plt.title("helium flux")
plt.xlabel("GeV/n")
plt.ylabel(r'Helium flux $[m^{2}\cdot s \cdot sr \cdot GeV/n]^{-1}$')

# BESS helium flux
df = pd.read_csv("/NAS/JC/paper/GRAMS_antihelium3_sensitivity/BESS_helium.csv")
BESS_helium = df.values.tolist()
BESS_helium = sorted(BESS_helium, key=lambda x: x[0])
BESS_helium_x = [row[0] for row in BESS_helium]
BESS_helium_y = [row[1] for row in BESS_helium]

# modify the energy convertion
BESS_helium_y = [y / (x ** 2.7) for x, y in zip(BESS_helium_x, BESS_helium_y)]
plt.plot(BESS_helium_x, BESS_helium_y, label = 'BESS helium result')

# AMS helium flux on year 2015 paper
df = pd.read_csv("/NAS/JC/paper/GRAMS_antihelium3_sensitivity/AMS_helium_2015.csv")
AMS_helium_2015 = df.values.tolist()
AMS_helium_2015 = sorted(AMS_helium_2015, key=lambda x: x[0])
AMS_helium_2015_x = [row[0] for row in AMS_helium_2015]
AMS_helium_2015_y = [row[1] for row in AMS_helium_2015]

# modify the energy convertion
AMS_helium_2015_y = [y / (x ** 2.7) for x, y in zip(AMS_helium_2015_x, AMS_helium_2015_y)]
plt.plot(AMS_helium_2015_x, AMS_helium_2015_y, label = 'AMS helium flux result 2015')

# AMS helium3 flux
df = pd.read_csv("/NAS/JC/paper/GRAMS_antihelium3_sensitivity/AMS_helium3.csv")
AMS_helium3 = df.values.tolist()
AMS_helium3 = sorted(AMS_helium3, key=lambda x: x[0])
AMS_helium3_x = np.array([row[0] for row in AMS_helium3])
# convert rigidity to kenetic energy per N
AMS_helium3_x = (np.sqrt(4 * AMS_helium3_x ** 2 + 2.808 ** 2)-2.808)/3
AMS_helium3_y = [row[1] for row in AMS_helium3]
# modify the energy convertion
AMS_helium3_y = [y / (x ** 2.7) for x, y in zip(AMS_helium3_x, AMS_helium3_y)]

# AMS helium4 flux
df = pd.read_csv("/NAS/JC/paper/GRAMS_antihelium3_sensitivity/AMS_helium4.csv")
AMS_helium4 = df.values.tolist()
AMS_helium4 = sorted(AMS_helium4, key=lambda x: x[0])
AMS_helium4_x = np.array([row[0] for row in AMS_helium4])
# convert rigidity to kenetic energy per N
AMS_helium4_x = (np.sqrt(4 * AMS_helium4_x ** 2 + 3.727 ** 2)-3.727)/4
AMS_helium4_y = [row[1] for row in AMS_helium4]
# modify the energy convertion
AMS_helium4_y = [y / (x ** 2.7) for x, y in zip(AMS_helium4_x, AMS_helium4_y)]

# creat helium flux curve fit
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel

# Assuming BESS_helium is already loaded as list of [energy, flux] pairs
x_data = np.array(AMS_helium_2015_x)
y_data = np.array(AMS_helium_2015_y)  # Flux

# Log-transform (essential for cosmic-ray spectra)
log_x = np.log10(x_data)
log_y = np.log10(y_data)

# Optimized Gaussian Process
kernel = ConstantKernel(1.0) * RBF(length_scale=0.5) 
gpr = GaussianProcessRegressor(
    kernel=kernel,
    alpha=0.1,                # Small noise term
    n_restarts_optimizer=10,  # Robust hyperparameter tuning
    random_state=42           # Reproducibility
)
gpr.fit(log_x.reshape(-1, 1), log_y)

def cal_helium_flux(energy, return_std=False):
    """
    Predicts helium flux with optional uncertainty.
    
    Args:
        energy (float/array): Energy in GeV/n
        return_std (bool): If True, returns (flux, uncertainty)
    
    Returns:
        Flux (and std if return_std=True)
    """
    energy = np.asarray(energy)
    log_energy = np.log10(energy).reshape(-1, 1)
    
    if return_std:
        log_flux, log_std = gpr.predict(log_energy, return_std=True)
        return 10**log_flux.flatten(), (10**log_flux * np.log(10) * log_std).flatten()
    return 10**gpr.predict(log_energy).flatten()
print("Flux is: ", cal_helium_flux(0.1))

# convert Nan Li ratio into flux
for i in range(len(DM_Nan_Li_upper_y)):
    DM_Nan_Li_upper_y[i] = DM_Nan_Li_upper_y[i] * cal_helium_flux(DM_Nan_Li_upper_x[i]) / ((DM_Nan_Li_upper_x[len(DM_Nan_Li_upper_x)-1] - DM_Nan_Li_upper_x[0])/len(DM_Nan_Li_upper_x))
for i in range(len(DM_Nan_Li_lower_y)):
    DM_Nan_Li_lower_y[i] = DM_Nan_Li_lower_y[i] * cal_helium_flux(DM_Nan_Li_lower_x[i]) / ((DM_Nan_Li_lower_x[len(DM_Nan_Li_lower_x)-1] - DM_Nan_Li_lower_x[0])/len(DM_Nan_Li_lower_x))


DM_Linden_y = [y_raw / x for x, y_raw in zip(DM_Linden_x, DM_Linden_y_raw)]

# Helium flux from AMS 2015
helium_flux_x = np.linspace(0.1, 120, 1000) 
helium_flux_converted = cal_helium_flux(helium_flux_x)

plt.legend()
```
