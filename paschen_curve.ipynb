```python
# data from https://www-eng.lbl.gov/~shuman/XENON/REFERENCES&OTHER_MISC/paschen_report.pdf
import math
import numpy as np
import matplotlib.pyplot as plt

# in air
gamma_se = 0.02
# torr^-1 cm^-1
A_air = 3.2
# V / (torr * cm)
B_air = 450
x = np.linspace(1.24, 2000, 50000)
# breakdown voltage in air
V_air = B_air * x / (np.log(A_air * x) - math.log(np.log(1 + 1 / gamma_se)))
plt.grid(True, alpha=0.3)
plt.loglog(x, V_air, label="Breakdown voltage in Air")
plt.ylabel("Breakdown Voltage(V)")
plt.xlabel("pressure x distance(torr x cm)")

# in zenon
gamma_se = 0.02
# torr^-1 cm^-1
A_ze = 2.7
# V / (torr * cm)
B_ze = 290
x = np.linspace(1.46, 2000, 50000)
# breakdown voltage in air
V_ze = B_ze * x / (np.log(A_ze * x) - math.log(np.log(1 + 1 / gamma_se)))
plt.grid(True, alpha=0.3)
plt.loglog(x, V_ze, label = "Breakdown voltage in Zenon")
plt.ylabel("Breakdown Voltage(V)")
plt.xlabel("pressure x distance(torr x cm)")


plt.axvline(x=760, color='red', linestyle='--', linewidth=2, label='1 atp(760 torr), 1cm distance')
plt.legend()

```
