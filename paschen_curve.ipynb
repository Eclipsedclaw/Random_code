```python
# data from https://www-eng.lbl.gov/~shuman/XENON/REFERENCES&OTHER_MISC/paschen_report.pdf
# gamma_se = 0.1 to 0.2 for metals,and gamma_se = 1 for some insulators. data from https://iopscience.iop.org/article/10.1088/1361-6595/ab094f/ampdf
import math
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

# in air
gamma_se = 0.02
# torr^-1 cm^-1
A_air = 3.2
# V / (torr * cm)
B_air = 450
x_air = np.linspace(1.23, 2000, 100000)
# breakdown voltage in air
V_air = B_air * x_air / (np.log(A_air * x_air) - math.log(np.log(1 + 1 / gamma_se)))
plt.loglog(x_air, V_air, label="Breakdown voltage in Air")

# in argon
gamma_se = 0.08
# torr^-1 cm^-1
A_ar = 11.5
# V / (torr * cm)
B_ar = 176
x_ar = np.linspace(0.22635, 2000, 100000)
# breakdown voltage in air
V_ar = B_ar * x_ar / (np.log(A_ar * x_ar) - math.log(np.log(1 + 1 / gamma_se)))
plt.loglog(x_ar, V_ar, label="Breakdown voltage in Argon")

# in zenon
gamma_se = 0.02
# torr^-1 cm^-1
A_ze = 24
# V / (torr * cm)
B_ze = 330
x_ze = np.linspace(0.16385, 2000, 100000)
# breakdown voltage in air
V_ze = B_ze * x_ze / (np.log(A_ze * x_ze) - math.log(np.log(1 + 1 / gamma_se)))
plt.loglog(x_ze, V_ze, label = "Breakdown voltage in Zenon")

plt.grid(True, alpha=0.3)
plt.ylabel("Breakdown Voltage(V)")
plt.xlabel("pressure x distance(torr x cm)")
plt.axvline(x=760, color='red', linestyle='--', linewidth=2, label='1 atp(760 torr), 1cm distance')
plt.axvline(x=3, color='darkred', linestyle='--', linewidth=2, label='3 torr, 1cm distance')
plt.legend()

```
