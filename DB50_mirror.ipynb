```python
import numpy as np
def channel_convert(channel):
    if channel < 18:
        changed_channel = 18 - channel
    elif channel < 34:
        changed_channel = 34 - channel + 17
    elif channel < 51:
        changed_channel = 51 - channel + 33
    return changed_channel

for n in range(50):
    print("DB50 channel "+str(n+1)+" mirror to "+str(channel_convert(n+1)))
```
